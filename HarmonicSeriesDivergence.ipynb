```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def harmonic_series(n_term, current):
    total = current + 1/n_term
    return total
```

```python
terms = np.linspace(1, 1000, 1000)
totals = []
total = 0
for term in terms:
    term = int(term)
    addition = harmonic_series(term, total)
    total += addition
    totals.append(total)
plt.plot(terms, totals)
plt.yscale("log")
plt.xscale("log")
```
