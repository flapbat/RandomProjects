```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("muted")
```

The number of pairs is $\displaystyle \frac{n\cdot (n-1)}{2}$

The probability that a single pair will have different birthdays is $\displaystyle \frac{364}{365}$

So the probability that all n people have different birthdays is $\displaystyle \big( \frac{364}{365} \big) ^ \text{number of pairs}$

Then because probability is normalized, the probability that there exists at least one matching pair is $1 - \displaystyle \big( \frac{364}{365} \big) ^ \text{number of pairs}$

```python
def find_prob(n):
    num_pairs = n * (n-1) / 2
    prob_one_mismatch = 1- 1/365
    prob_mismatch = prob_one_mismatch**num_pairs
    prob_match = 1-prob_mismatch
    return prob_match
print(find_prob(23))
```

```python
ns = np.linspace(0,50, 51)
probs = find_prob(ns)
plt.plot(ns, probs)
plt.xlabel("Number of People")
plt.ylabel("Probability of Match")
plt.plot([23,23],[0,1], color = "red")
plt.annotate("23", xy = (24,0), color = "red")
plt.plot()
```

```python
def find_prob_mis(n):
    num_pairs = n * (n-1) / 2
    prob_one_mismatch = 1- 1/365
    prob_mismatch = prob_one_mismatch**num_pairs
    return prob_mismatch
```

```python
find_prob_mis(23)
```

```python
ns = np.linspace(0,50, 51)
probs_mis = find_prob_mis(ns)
plt.plot(ns, probs_mis, label = "No Pairs Match")
plt.plot(ns, probs, label = "At Least One Pair Matches")
plt.xlabel("Number of People")
plt.ylabel("Probability")
plt.plot([23,23],[0,1], color = "red")
plt.annotate("23", xy = (24,0), color = "red")
plt.legend(loc = "upper right")
plt.plot()
```
