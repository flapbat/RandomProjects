```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_prob(n):
    num_pairs = n * (n-1) / 2
    prob_one_mismatch = 1- 1/365
    prob_mismatch = prob_one_mismatch**num_pairs
    prob_match = 1-prob_mismatch
    return prob_match
print(find_prob(23))
```

```python
ns = np.linspace(0,50, 51)
probs = find_prob(ns)
plt.plot(ns, probs)
plt.xlabel("Number of People")
plt.ylabel("Probability of Match")
plt.plot([23,23],[0,1], color = "red")
plt.annotate("23", xy = (24,0), color = "red")
plt.plot()
```

```python
ns = np.linspace(0,50, 51)
num_prs = ns * (ns-1) / 2
plt.plot(ns, num_prs)
plt.yscale("log")
plt.xlabel("Number of People")
plt.ylabel("Number of Pairs")
plt.plot([23,23],[10**0,10**3], color = "red")
plt.annotate("23", xy = (24,1), color = "red")
plt.plot()
```

```python
def find_prob_mis(n):
    num_pairs = n * (n-1) / 2
    prob_one_mismatch = 1- 1/365
    prob_mismatch = prob_one_mismatch**num_pairs
    return prob_mismatch
```

```python
ns = np.linspace(0,50, 51)
probs_mis = find_prob_mis(ns)
plt.plot(ns, probs_mis)
plt.xlabel("Number of People")
plt.ylabel("Probability of Mismatch")
plt.plot([23,23],[0,1], color = "red")
plt.annotate("23", xy = (24,0), color = "red")
plt.plot()
```
